# Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Creating new column "max_order"
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking for new column
ords_prods_merge.head(15)
```

```python
# Stablish a loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag in dataframe
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(10)
```

# Analyze spending habits across the three customer types using the loyalty flag, and check basic statistics of product prices for each category (Loyal, Regular, and New)

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

# Create a spending flag for each user based on their average purchase price. If it's less than 10, label them as "Low spender"; if it's 10 or higher, as "High spender".

```python
# Creating column for average_product_price
ords_prods_merge['avg_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking the new column
ords_prods_merge[['user_id', 'avg_product_price']].head(20)
```

```python
# Stablish a flag for low and high spender
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_product_price']<10,'spender_flag']='Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_product_price']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Check new flag column
ords_prods_merge[['user_id', 'avg_product_price', 'spender_flag']].head(20)
```

# Create an order frequency flag based on the median of "days_since_prior_order". Classify customers as "Non-frequent", "Regular", or "Frequent" based on the median.

```python
# Creating column for avg_days_since_prior_order
ords_prods_merge['users_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
# Checking the new column
ords_prods_merge[['user_id', 'days_since_prior_order', 'users_frequency']].head(20)
```

```python
# Stablish a flag for users_frecuency
ords_prods_merge.loc[ords_prods_merge['users_frequency'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['users_frequency']>10) & (ords_prods_merge['users_frequency']<=20),'order_frequency'] = 'Regular customer'
                     
```

```python
ords_prods_merge.loc[ords_prods_merge['users_frequency'] <= 10, 'order_frequency'] = 'Frequent Customer'
```

```python
# Check order frequency
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
# Check new flag column
ords_prods_merge[['user_id', 'users_frequency', 'order_frequency']].head(20)
```
